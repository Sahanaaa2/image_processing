```python
#Python program to implement Histogram(using plot)
import cv2
import matplotlib.pyplot as plt
import numpy as np
# Read the image in grayscale
img = cv2.imread('images\\bird.jpg', 0)
row,col = img.shape

# Calculate the histogram
histogram = np.zeros(256)
for i in range(row):
    for j in range(col):
        intensity = img[i, j]
        histogram[intensity] += 1
        
hist1 = cv2.calcHist([img], [0], None, [256], [0, 256])
# Plot the original image and the histogram
plt.figure(figsize=(12, 5))

# Plot the original image
plt.subplot(1, 3, 1)
plt.title("Original Image")
plt.imshow(img, cmap='gray')
plt.axis('off')  # Hide the axis

# Plot the histogram (without built in function)
plt.subplot(1, 3, 2)
plt.title("Histogram Representation\n(without built in function)")
plt.plot(histogram,color="maroon")
plt.xlim([0, 255])
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')

# Plot the histogram (with built in function)
plt.subplot(1, 3, 3)
plt.title("Histogram Representation\n(with built in function)")
plt.plot(hist1,color="green")
plt.xlim([0, 255])
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')

# Show the plots
plt.tight_layout()
plt.show()
```

```python
#Python program to implement average smoothing
import cv2
import numpy as np
import matplotlib.pyplot as plt


def average_smooth(img,kernel_size):
    rows,cols=img.shape
    modified_img=np.zeros((rows,cols))
    for i in range(rows):
        for j in range(cols):
            sum_1=0
            try:
                for p in range(-(kernel_size//2),kernel_size//2+1):
                    for q in range(-(kernel_size//2),kernel_size//2+1):
                        sum_1+=img[(i+p,j+q)]
            except IndexError as e:
                pass
            modified_img[i,j]=sum_1/(kernel_size**2)
    return modified_img


img=input("enter image::")

# Kernel size for smoothing
kernel_size = int(input("enter kernel size::"))

if img is None:
    print("Error:Unable to load the image")
else:
# Load the image in grayscale mode
    img = cv2.imread(img, cv2.IMREAD_GRAYSCALE)


# Apply average smoothing using cv2.blur
    smoothed_image1 = cv2.blur(img, (kernel_size, kernel_size))
   
# Display the original and smoothed images
    plt.figure(figsize=(12, 8))

# Original image
    plt.subplot(1, 3, 1)
    plt.title("Original Image")
    plt.imshow(img, cmap='gray')

# Smoothed image
    plt.subplot(1, 3, 2)
    plt.title("Smoothed Image(with built in)")
    plt.imshow(smoothed_image1, cmap='gray')

    smooth_img2=average_smooth(img,kernel_size)

    plt.subplot(1, 3, 3)
    plt.title("Smoothed image(without built in) ")
    plt.imshow(smooth_img2,cmap='gray')

# Show the plots
    plt.show()
```
