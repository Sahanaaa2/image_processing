# Global Thresholding(Binary Image)

```python
#Python program to implement global thresholding with built in functions

#import libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the grayscale image
image = cv2.imread('cat.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
# Check if the image was loaded correctly
if image is None:
    raise ValueError("Image not found or path is incorrect")
```

```python
# Set the threshold value
threshold_value = 125 # You can adjust this value as needed
```

```python
# Perform global thresholding
_, binary_image = cv2.threshold(image, threshold_value, 255, cv2.THRESH_BINARY)
```

```python
# Display the original and segmented images
plt.figure(figsize=[10,5])

plt.subplot(1, 2, 1)
plt.title("Original Grayscale Image")
plt.imshow(image, cmap='gray')

plt.subplot(1, 2, 2)
plt.title("Binary Image after Thresholding")
plt.imshow(binary_image, cmap='gray')

plt.show()
```

```python
#Python program to implement global thresholding without built in functions

#import libraries
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
#function to perform global thresholding
def global_thresholding(image, threshold_value):
    binary_image = np.zeros_like(image)
    height, width = image.shape
    
    for i in range(height):
        for j in range(width):
            if image[i, j] > threshold_value:
                binary_image[i, j] = 255
            else:
                binary_image[i, j] = 0
    
    return binary_image
```

```python
#specify the image name and threshold value
image_path = 'drop.jpg'
manual_threshold = 127
```

```python
# Load the image
original_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
```

```python
# Apply global thresholding
binary_image = global_thresholding(original_image, manual_threshold)
```

```python
#plot the image
plt.figure(figsize=[10, 5])
    
plt.subplot(1, 2, 1)
plt.title("Original Grayscale Image")
plt.imshow(original_image, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title("Binary Image after Thresholding")
plt.imshow(binary_image, cmap='gray')
plt.axis('off')
```

# Local Thresholding

```python
#Python program to implement local thresholding without built in functions

#import libraries
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
#Calculate the local threshold for a block of the image
def calculate_local_threshold(block):
    return np.mean(block)  # Use the mean intensity value of the block as the local threshold
```

```python
#function to perform local thresholding
def local_thresholding(image, block_size):
    height, width = image.shape
    print(image.shape)
    binary_image = np.zeros_like(image)
    
    for i in range(0, height, block_size):
        for j in range(0, width, block_size):
            block = image[i:i+block_size, j:j+block_size]
            local_thresh = calculate_local_threshold(block)
            
            for x in range(block.shape[0]):
                for y in range(block.shape[1]):
                    if block[x, y] > local_thresh:
                        binary_image[i + x, j + y] = 255
                    else:
                        binary_image[i + x, j + y] = 0
    
    return binary_image
```

```python
#specify the image
image_path = 'white.jpg'
```

```python
# Block size for local thresholding
block_size = 16  # You can adjust this value as needed
```

```python
# Load the image
original_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
```

```python
# Apply local thresholding
binary_image = local_thresholding(original_image, block_size)
```

```python
#plotting the images
plt.figure(figsize=[10, 5])
    
plt.subplot(1, 2, 1)
plt.title("Original Grayscale Image")
plt.imshow(original_image, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title("Binary Image after Local Thresholding")
plt.imshow(binary_image, cmap='gray')
plt.axis('off')
```

# Sobel Edge Detection

```python
#Python program to implement edge detection by Sobel Edge Detection method

#import libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the image in grayscale
image = cv2.imread('C:\\Users\\User\\image_processing\\images\\cats.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
# Apply Sobel edge detection
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)
sobel_combined = np.sqrt(sobel_x**2 + sobel_y**2)
```

```python
# Display the original and Sobel edge-detected images
plt.figure(figsize=[10, 5])
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(sobel_combined, cmap='gray')
plt.title('Sobel Edge-detected Image')
plt.axis('off')

plt.show()
```

# Prewitt Edge Detection

```python
#Python program to implement edge detection by Prewitt Edge Detection method


#import libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the image in grayscale
image = cv2.imread('flower.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
# Define the Prewitt kernels
kernel_x = np.array([[-1, 0, 1],
                     [-1, 0, 1],
                     [-1, 0, 1]])

kernel_y = np.array([[-1, -1, -1],
                     [0, 0, 0],
                     [1, 1, 1]])
```

```python
# Apply convolution with the Prewitt kernels
prewitt_x = cv2.filter2D(image, -1, kernel_x)
prewitt_y = cv2.filter2D(image, -1, kernel_y)
```

```python
# Calculate the magnitude of gradients
prewitt_combined = np.sqrt(prewitt_x**2 + prewitt_y**2)
```

```python
# Display the original and Prewitt edge-detected images
plt.figure(figsize=[10, 5])
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(prewitt_combined, cmap='gray')
plt.title('Prewitt Edge-detected Image')
plt.axis('off')

plt.show()
```

# Roberts Cross Edge Detection

# Zero-crossing Edge Detection

```python
#Python program to implement edge detection by Zero Crossing Edge Detection method

#import libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the image in grayscale
image = cv2.imread('flower.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
# Apply Laplacian edge detection
laplacian = cv2.Laplacian(image, cv2.CV_64F)
```

```python
# Find zero-crossings in the Laplacian image
rows, cols = laplacian.shape
zero_crossings = np.zeros_like(laplacian, dtype=np.uint8)
for i in range(1, rows-1):
    for j in range(1, cols-1):
        neighbors = [laplacian[i-1, j], laplacian[i+1, j], laplacian[i, j-1], laplacian[i, j+1],
                     laplacian[i-1, j-1], laplacian[i-1, j+1], laplacian[i+1, j-1], laplacian[i+1, j+1]]
        if np.any(np.multiply(neighbors, laplacian[i, j]) < 0):
            zero_crossings[i, j] = 255
```

```python
# Display the original and zero-crossing edge-detected images
plt.figure(figsize=[10, 5])
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(zero_crossings, cmap='gray')
plt.title('Zero-crossing Edge-detected Image')
plt.axis('off')

plt.show()
```

# Segmentation

```python
#Python program to implement segmentation

#import libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the image in grayscale
image = cv2.imread('flower.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
# Reshape the image into a 2D array of pixels
pixels = image.reshape((-1, 1))
```

```python
# Convert to float32
pixels = np.float32(pixels)
```

```python
# Define criteria and apply k-means clustering
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 0.2)
k = 5  # Number of clusters
_, labels, centers = cv2.kmeans(pixels, k, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
```

```python
# Convert back to 8-bit values
centers = np.uint8(centers)
```

```python
# Map the labels to their respective centers
segmented_image = centers[labels.flatten()]
```

```python
# Reshape back to the original image dimensions
segmented_image = segmented_image.reshape(image.shape)
```

```python
# Display the original and segmented images
plt.figure(figsize=[10, 5])
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(segmented_image, cmap='gray')
plt.title('Segmented Image')
plt.axis('off')

plt.show()
```
