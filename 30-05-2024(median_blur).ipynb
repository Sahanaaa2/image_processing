```python
#Python program to median filter without using built in functions
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

def median_filter(image, kernel_size):
    height, width, channels = image.shape
    output = np.zeros((height, width, channels), dtype=np.uint8)
    k_half = kernel_size // 2
    
    for i in range(k_half, height - k_half):
        for j in range(k_half, width - k_half):
            for c in range(channels):
                neighbors = []
                for m in range(-k_half, k_half + 1):
                    for n in range(-k_half, k_half + 1):
                        neighbors.append(image[i+m, j+n, c])
                output[i, j, c] = np.median(neighbors)
    
    return output

# Load the image
image_path = 'cat.jpg'
image = np.array(Image.open(image_path))

# Remove noise using median filter
denoised_median = median_filter(image, kernel_size=3)

# Display the original and denoised images
plt.figure(figsize=(15, 5))
plt.subplot(121)
plt.imshow(image)
plt.title('Original Image')
plt.axis('off')

plt.subplot(122)
plt.imshow(denoised_median)
plt.title('Median Blurred Image')
plt.axis('off')

plt.show()
```

```python
#Python program to implement median blur of 5 images without built in functions
import cv2
import numpy as np

# Load the five images
img1 = cv2.imread('images\\bird.jpg')
img2 = cv2.imread('images/cats.jpg')
img3 = cv2.imread('images/green.jpg')
img4 = cv2.imread('images\lion.jpg')
img5 = cv2.imread('images\\nature.jpeg')

# Get the dimensions of the first image
height, width, channels = img1.shape

# Resize the images to match the dimensions of the first image
img2 = cv2.resize(img2, (width, height))
img3 = cv2.resize(img3, (width, height))
img4 = cv2.resize(img4, (width, height))
img5 = cv2.resize(img5, (width, height))

# Convert images to numpy arrays
images = [img1, img2, img3, img4, img5]
images_array = np.array(images)

# Calculate the average pixel values
average_image = np.zeros((height, width, channels), dtype=np.uint8)
for i in range(height):
    for j in range(width):
        for k in range(channels):
            sum = 0
            for img in images_array:
                sum += img[i, j, k]
            average_image[i, j, k] = sum // len(images_array)

# Display or save the average image
plt.figure(figsize=(15, 5))
plt.subplot(121)
plt.imshow(average_image)
plt.title('Filtered Image')
plt.axis('off')
```

```python
import matplotlib.pyplot as plt
import cv2

# Load the images
image1 = cv2.imread('image1.jpg')
image2 = cv2.imread('image2.jpg')

# Check if the images are loaded successfully
if image1 is None or image2 is None:
    print("Error: Could not read one or more images.")
    exit()

# Ensure both images have the same dimensions
if image1.shape != image2.shape:
    print("Error: Images must have the same dimensions.")
    exit()

# Subtract one image from the other
subtracted_image = cv2.subtract(image1, image2)

# Plot the images using subplots
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Plot image 1
axes[0].imshow(cv2.cvtColor(image1, cv2.COLOR_BGR2RGB))
axes[0].set_title('Image 1')
axes[0].axis('off')

# Plot image 2
axes[1].imshow(cv2.cvtColor(image2, cv2.COLOR_BGR2RGB))
axes[1].set_title('Image 2')
axes[1].axis('off')

# Plot subtracted image
axes[2].imshow(cv2.cvtColor(subtracted_image, cv2.COLOR_BGR2RGB))
axes[2].set_title('Subtracted Image')
axes[2].axis('off')

plt.show()
```
