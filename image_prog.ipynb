```python
# Program to display image in pixel manipulation using cv2 
import cv2 
img = cv2.imread('C:/Users/CS/Downloads/lion.jpg')  # Load the input image 
(row, col) = img.shape[0:2]   # Obtain the dimensions of the image array using the shape method 
for i in range(row):          # Take the average of pixel values of the BGR Channels to convert the colored image to grayscale image 
    for j in range(col): 
        img[i, j] = sum(img[i, j]) * 0.33   # Find the average of the BGR pixel values 
cv2.imshow('Grayscale Image', img) 
cv2.waitKey(0) 
cv2.destroyAllWindows()     # Window shown waits for any key pressing event 



#Output displayed as task bar icon(1 image)
```

```python
# Program to resize an existing image                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                     
import cv2
import numpy as np
import matplotlib.pyplot as plt
# Load the image
image =cv2.imread('C:/Users/CS/Downloads/eye.jpg')
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR image to RGB
scale_factor_1 = 3.0     # Define the scale factor,Increase the size by 3 times
# Decrease the size by 3 times
scale_factor_2 = 1/3.0
height, width = image_rgb.shape[:2]   # Get the original image dimensions
new_height = int(height * scale_factor_1)  # Calculate the new image dimensions
new_width = int(width * scale_factor_1)
zoomed_image = cv2.resize(src =image_rgb,          # Resize the image
                          dsize=(new_width, new_height), 
                          interpolation=cv2.INTER_CUBIC)
new_height1 = int(height * scale_factor_2)    # Calculate the new image dimensions
new_width1 = int(width * scale_factor_2)
scaled_image = cv2.resize(src= image_rgb,      # Scaled image
                          dsize =(new_width1, new_height1), 
                          interpolation=cv2.INTER_AREA)
fig, axs = plt.subplots(1, 3, figsize=(10, 4))   # Create subplots
axs[0].imshow(image_rgb)         # Plot the original image
axs[0].set_title('Original Image Shape:'+str(image_rgb.shape))
axs[1].imshow(zoomed_image)       # Plot the Zoomed Image
axs[1].set_title('Zoomed Image Shape:'+str(zoomed_image.shape))
axs[2].imshow(scaled_image)   # Plot the Scaled Image
axs[2].set_title('Scaled Image Shape:'+str(scaled_image.shape))
for ax in axs:                  # Remove ticks from the subplots
    ax.set_xticks([])
    ax.set_yticks([])
plt.tight_layout()                 # Display the subplots
plt.show()
```

```python
#Program to display image in normal form
import cv2
import numpy as np
import matplotlib.pyplot as plt
image = cv2.imread('C:/Users/CS/Downloads/monument.jpg')       # Load the image
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)   # Convert BGR image to RGB
b, g, r = cv2.split(image_rgb)          # Split the image into channels
min_value = 0                       # Normalization parameter
max_value = 1
norm_type = cv2.NORM_MINMAX
# Normalize each channel
b_normalized = cv2.normalize(b.astype('float'), None, min_value, max_value, norm_type)
g_normalized = cv2.normalize(g.astype('float'), None, min_value, max_value, norm_type)
r_normalized = cv2.normalize(r.astype('float'), None, min_value, max_value, norm_type)

# Merge the normalized channels back into an image
normalized_image = cv2.merge((b_normalized, g_normalized, r_normalized))
print(normalized_image[:,:,0])     # Normalized image
plt.imshow(normalized_image)
plt.xticks([])
plt.yticks([])
plt.title('Normalized Image')
plt.show()



#Output brightness will vary accordingly with min and max values
```

```python
#Program to translate an image
import numpy as np
import cv2 as cv
img = cv2.imread("C:/Users/CS/Downloads/dog.jpg",0)
rows, cols = img.shape
M = np.float32([[1, 0, 100], [0, 1, 50]])
dst = cv.warpAffine(img, M, (cols, rows))
cv.imshow('img', dst)
cv.waitKey(0)
cv.destroyAllWindows()

#Image in the output is slightly changed with its position(task bar icon)
```

```python
#Program to display image by reflecting it
import numpy as np
import cv2 as cv
img = cv2.imread("C:/Users/CS/Downloads/micky.png",0)
rows, cols = img.shape
M = np.float32([[1, 0, 0],[0, -1, rows],[0, 0, 1]])
reflected_img = cv.warpPerspective(img, M,(int(cols),int(rows)))
cv.imshow('img', reflected_img)
cv.imwrite('reflection_out.jpg', reflected_img)
cv.waitKey(0)
cv.destroyAllWindows()



#Output image is displayed as task bar icon
```
