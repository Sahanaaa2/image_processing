```python
#Program to display original image,negative image,binary image,red,green,blue image
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('flower.jpg')
plt.subplot(2,3,1)
plt.title("Original")
plt.imshow(image)

gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.subplot(2,3,2)
plt.title("Grayscale")
plt.imshow(gray_image, cmap = 'gray')

_, binary_image = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)
plt.subplot(2,3,3)
plt.title("Binary")
plt.imshow(binary_image, cmap = 'gray')

blue_channel, green_channel, red_channel = cv2.split(image)

plt.subplot(2,3,4)
plt.title('Red Channel')
plt.imshow(red_channel, cmap = 'Reds')

plt.subplot(2,3,5)
plt.title('Green Channel')
plt.imshow(green_channel, cmap='Greens')

plt.subplot(2,3,6)
plt.title('Blue Channel')
plt.imshow(blue_channel, cmap='Blues')
```

```python
#Program to display mean,standard deviation,corelation and co variance of an image
import cv2
import numpy as np
img = cv2.imread("droplet.jpg", cv2.IMREAD_GRAYSCALE)                    # Read the image
mean_value = np.mean(img)                                        # Calculate mean and standard deviation
std_deviation = np.std(img)
correlation_coefficient = np.corrcoef(img)[0, 1]                # Calculate correlation coefficient
print("Mean: {:.2f}".format(mean_value))                        # Display results
print("Standard Deviation: {:.2f}".format(std_deviation))
print("Correlation Coefficient: {:.2f}".format(correlation_coefficient))
```

```python
#Program to display bit planes of an image
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the image
img = cv2.imread("stone.jpg", cv2.IMREAD_GRAYSCALE)

# Iterate through each bit plane
for i in range(8):
    # Extract the i-th bit plane
    bit_plane = (img >> i) & 1

    # Display the bit plane
    plt.subplot(2, 4, i+1)
    plt.imshow(bit_plane, cmap='gray')
    plt.title(f'Bit Plane {i}')
    plt.axis('off')

# Adjust layout and display
plt.tight_layout()
plt.show()
```

```python
# Program to resize an existing image                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                     
import cv2
import numpy as np
import matplotlib.pyplot as plt
# Load the image
image =cv2.imread('eyes.jpg')
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR image to RGB
scale_factor_1 = 3.0     # Define the scale factor,Increase the size by 3 times
# Decrease the size by 3 times
scale_factor_2 = 1/3.0
height, width = image_rgb.shape[:2]   # Get the original image dimensions
new_height = int(height * scale_factor_1)  # Calculate the new image dimensions
new_width = int(width * scale_factor_1)
zoomed_image = cv2.resize(src =image_rgb,          # Resize the image
                          dsize=(new_width, new_height), 
                          interpolation=cv2.INTER_CUBIC)
new_height1 = int(height * scale_factor_2)    # Calculate the new image dimensions
new_width1 = int(width * scale_factor_2)
scaled_image = cv2.resize(src= image_rgb,      # Scaled image
                          dsize =(new_width1, new_height1), 
                          interpolation=cv2.INTER_AREA)
fig, axs = plt.subplots(1, 3, figsize=(10, 4))   # Create subplots
axs[0].imshow(image_rgb)         # Plot the original image
axs[0].set_title('Original Image Shape:'+str(image_rgb.shape))
axs[1].imshow(zoomed_image)       # Plot the Zoomed Image
axs[1].set_title('Zoomed Image Shape:'+str(zoomed_image.shape))
axs[2].imshow(scaled_image)   # Plot the Scaled Image
axs[2].set_title('Scaled Image Shape:'+str(scaled_image.shape))
for ax in axs:                  # Remove ticks from the subplots
    ax.set_xticks([])
    ax.set_yticks([])
plt.tight_layout()                 # Display the subplots
plt.show()
```

```python
#Program to implement image intensity,slicing techniques for image enhancement 
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread("cat.jpg", cv2.IMREAD_GRAYSCALE)            # Read the image
low_threshold = 100  
high_threshold = 200
result_img = np.zeros_like(img)                                   # Apply intensity slicing
result_img[(img >= low_threshold) & (img <= high_threshold)] = 255
plt.figure(figsize=(10, 5))                         # Plot the original and result images using Matplotlib
plt.subplot(1, 2, 1)                               # Original image
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')
plt.subplot(1, 2, 2)                                # Enhanced image
plt.imshow(result_img, cmap='gray')
plt.title('Enhanced Image')
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
#Program to display an image using median blurr
import cv2
import matplotlib.pyplot as plt
import numpy as np               
image = cv2.imread('dill.jpg')                     # Load the image
plt.subplot(1, 2, 1)                              
plt.subplot(1, 2, 1)                              
plt.title("Original")
plt.imshow(image)
filtered_image = cv2.medianBlur(image, 11)          # Remove noise using a median filter
cv2.imwrite('Median Blur.jpg', filtered_image)     #Save the image
plt.subplot(1, 2, 2)                               #Plot the blured image
plt.title("Median Blur")
plt.imshow(filtered_image)
plt.show()
```
