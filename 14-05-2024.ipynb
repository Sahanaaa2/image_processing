```python
#Program to obtain row major representation of a given matrix
import cv2
img=cv2.imread("road.jpg",0)
rows,cols=img.shape
vector=[ ]
for i in range(rows):
    for j in range(cols):
        vector.append(img[i,j])
print("Row Major representation: ",vector)
print("Number of rows: ",rows)
print("Number of colums: ",cols)
```

```python
#Program to obtain Eucledian Distance between given two vectors
import math
v1 = [78, 77, 88]
v2 = [23, 33, 22]
sum1 = 0
for i in range(len(v1)):
    sum1 += ( v1[i] - v2[i] ) ** 2
ED=math.sqrt(sum1)
print("Euclidean Distance: ",ED)
```

```python
#Program to display Euclidean Distance of given multiple images without using built in functions
import math
import cv2

img1 = cv2.imread("flower.jpg", 0)
img2 = cv2.imread("cat.jpg", 0)

rows1,cols1 = img1.shape
rows2, cols2 = img2.shape

print(rows1)
print(cols1)

print(rows2)
print(cols2)

sum1 = 0
for i in range(rows1):
    for j in range(cols1):
        sum1 += (img1[i,j] - img2[i,j]) ** 2
ED = math.sqrt(sum1)

print("Eucidean distance :: ",ED)
```

```python
#Python program to find the average of grayscale image by using its vector without using built in functions
def average(image):
    sum = 0
    count = 0
    for row in image:
        for pixel in row:
            sum += pixel
            count += 1
    average = sum / count
    return average
image = [
    [90, 20, 30],
    [40, 50, 60],
    [70, 80, 90]
]
average = average(image)
print("Average intensity of the grayscale image:", average)
```

```python
#Python program to find the variance of grayscale image by using vector value without using built in functions
def variance(image):
    sums= 0
    sum_squared = 0
    count = 0
    for row in image:
        for pixel in row:
            sums += pixel
            sum_squared += pixel ** 2
            count += 1
    mean = sums / count
    variance = (sums / count) - (mean ** 2)
    return variance
image = [
    [10, 20, 30],
    [40, 50, 60],
    [70, 80, 90]
]
variance = variance(image)
print("Variance of the grayscale image:", variance)
```

```python
#Python program to find the variance of grayscale image by using vector value without using built in functions
import cv2
S = [[2, 5, 3],
    [6, 1, 8],
    [4, 7, 8]]
print("Original matrix\n",S)
sum1 = 0
for i in range(3):
    for j in range(3):
        sum1 += S[i][j]
mean = sum1 / 9
print("Mean :: ", mean)
varianceSum = 0
for i in range(3):
    for j in range(3):
        varianceSum += (S[i][j] - mean) ** 2
variance = varianceSum / 9
print("Variance :: ", variance)
```

```python
#Python program to find the mean,variance of grayscale image without using built in functions
import cv2
img = cv2.imread("flower.jpg", 0)
print("Original matrix\n",img)
rows, cols = img.shape
sum1 = 0
for i in range(rows):
    for j in range(cols):
        sum1 += img[i,j]
mean = sum1 / (rows * cols)
print("Mean :: ", mean)
varianceSum = 0
for i in range(rows):
    for j in range(cols):
        varianceSum += (img[i,j] - mean) ** 2
variance = varianceSum / (rows * cols)
print("Variance :: ", variance)
```

```python
#Program to plot mean,variance of 20 images without using built in functions
import cv2
import numpy as np
import matplotlib.pyplot as plt
img=input("Enter the image: ")
image = cv2.imread(img,0)
image_array = np.array(image)
rows, cols = image.shape
sum1 = 0
for i in range(rows):
    for j in range(cols):
        sum1 += image[i,j]
mean = sum1 / (rows * cols)
print("Mean :: ", mean)
varianceSum = 0
for i in range(rows):
    for j in range(cols):
        varianceSum += (image[i,j] - mean) ** 2
variance = varianceSum / (rows * cols)
print("Variance :: ", variance)
plt.figure(figsize=(10, 5))

# Plotting the image
plt.subplot(1, 2, 1)
plt.imshow(image_array, cmap='gray')
plt.title('Image')

# Plotting the mean and variance
plt.subplot(1, 2, 2)
plt.bar(['Mean', 'Variance'], [mean, variance])
plt.title('Mean and Variance')
plt.ylabel('Value')
```
